```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
from functools import partial

# Create 1,512 tensor filled with the values of a bell curve

# Create a 1x512x1 tensor
gaussian = torch.zeros(1, 1, 512)

# Create a range from 0 to 511
x = torch.arange(0, 512)

# Normalize and center the range around 0
x = (x - 256) / 256  # Adjust this division for different spread

# Mean and standard deviation for the Gaussian
mu = 0
sigma = 1 / 3  # Adjust sigma for different spread

# Apply the Gaussian function
gauss = torch.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

# Replace the tensor values
gaussian[0, 0, :] = gauss

# gaussian_vals = np.tile(gaussian_vals.detach().numpy(), (1, 1, 2))
gaussian_vals = torch.tile(gaussian, (1, 1, 10))
gaussian_vals[0, 0, 400:600] /= 10000
gaussian_vals.requires_grad = True

probs = torch.softmax(gaussian_vals, dim=-1)
probs_copy = probs.clone()

log_probs = torch.log(probs)

plt.plot(probs[0, 0, :].detach().numpy())
plt.show()

# Create a same shape tensor with random probabilities
rand_vals = torch.tile(gaussian, (1, 1, 40))
rand_vals = rand_vals[:, :, ::4]
rand_vals.requires_grad = True

rand_vals_copy = rand_vals.clone()

rand_probs = torch.softmax(rand_vals, dim=-1)
rand_copy = rand_probs.clone()

log_rand_probs = torch.log(rand_probs)

plt.plot(rand_probs[0, 0, :].detach().numpy())
plt.show()

from scipy import stats
from equivariant_pose_graph.utils.loss_utils import js_div, js_div_mod, wasserstein_distance
fns = [partial(F.mse_loss, reduction='mean'), wasserstein_distance, js_div, partial(js_div_mod, d_2=1), partial(js_div_mod, d_2=0.1), partial(js_div_mod, d_2=0.01), partial(js_div_mod, d_2=0.001)]

# Use KL divergence to bring the random tensor closer to the Gaussian
for fn_idx, fn in enumerate(fns):
    print(f'Function: {fn}')
    rand_vals = torch.tile(gaussian, (1, 1, 40))
    rand_vals = rand_vals[:, :, ::4]/2
    rand_vals.requires_grad = True

    print(f'rand_vals shape: {rand_vals.shape}')
    print(f'gaussian_vals shape: {gaussian_vals.shape}')

    rand_probs_orig = torch.softmax(rand_vals, dim=-1)    
    
    for i in range(10000):
        if fn_idx == 0:
            loss = fn(F.log_softmax(gaussian_vals, dim=-1), F.log_softmax(rand_vals, dim=-1))
        elif fn_idx == 1:
            loss = fn(gaussian_vals, rand_vals)
            vec_list = np.aranage(gaussian_vals.shape[-1])
        else:
            loss = fn(p=F.log_softmax(gaussian_vals, dim=-1), q=F.log_softmax(rand_vals, dim=-1), reduction='batchmean')
        # loss_2 = fn(p=F.log_softmax(rand_vals, dim=-1), q=F.log_softmax(gaussian_vals, dim=-1), reduction='sum', log_target=True)
        
        loss.backward()
        if i < 10:
            print(f'rand_vals.grad: {rand_vals.grad}')
        rand_vals.data -= 5*rand_vals.grad
        rand_vals.grad.zero_()
        rand_probs = torch.softmax(rand_vals, dim=-1)
        log_rand_probs = torch.log(rand_probs)
    
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    ax.plot(probs[0, 0, :].detach().numpy())
    ax.plot(rand_probs_orig[0, 0, :].detach().numpy())
    ax.plot(rand_probs[0, 0, :].detach().numpy())
    plt.show()
    


```
